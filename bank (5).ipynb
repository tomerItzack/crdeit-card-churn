# התפלגות הלקוחות הפורשים


```python
colors = ['purple', 'blue', 'red', 'green', 'gray', 'orange','brown']  

value_counts = df['Income'].value_counts()
percentages = (value_counts / len(df['Income'])) * 100

# Sort the percentages
sorted_percentages = percentages.sort_values()

# Ensure the colors list matches the number of unique values
sorted_colors = colors[:len(sorted_percentages)]

# Plot the percentages
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust the figsize as needed
bars = ax.bar(sorted_percentages.index, sorted_percentages, color=sorted_colors)

# Add title and labels
plt.title('non customers')
plt.xlabel('Income')
plt.ylabel('Percentage')

# Format the y-axis to show the percentage sign
formatter = FuncFormatter(lambda y, _: '{:.0f}%'.format(y))
ax.yaxis.set_major_formatter(formatter)

# Add percentage labels on the bars
for bar, percentage in zip(bars, sorted_percentages):
    height = bar.get_height()
    ax.annotate(f'{percentage:.1f}%',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords='offset points',
                ha='center', va='bottom')

# Show the plot
plt.show()
```

# אחוז הלקוחות הפורשים לפי הכנסה

```python
# Create separate dataframes for attrited and existing customers
attrited_df = dist[dist['Customer_Status'] == 'Attrited Customer']
existing_df = dist[dist['Customer_Status'] == 'Existing Customer']

# Count the number of customers in each income category
attrited_counts = attrited_df['Income'].value_counts()
existing_counts = existing_df['Income'].value_counts()

# Combine the counts into a single DataFrame
df_Income = pd.DataFrame({
    'Attrited_Customers': attrited_counts,
    'Existing_Customers': existing_counts
}).fillna(0)

# Calculate total customers and percentage of attrited customers
df_Income['Total_Customers'] = df_Income['Attrited_Customers'] + df_Income['Existing_Customers']
df_Income['Percentage_Attrited'] = (df_Income['Attrited_Customers'] / df_Income['Total_Customers']) * 100

# Plotting
plt.figure(figsize=(10, 6))

# Plotting the percentage of attrited customers by income category
bars = plt.bar(df_Income.index, df_Income['Percentage_Attrited'], color=colors)

# Adding percentage numbers above each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{yval:.2f}%', ha='center', va='bottom')

plt.xlabel('Income Category')
plt.ylabel('Percentage of Attrited Customers (%)')
plt.title('Percentage of Attrited Customers by Income Category')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, max(df_Income['Percentage_Attrited']) + 5)  # Adding some buffer to the ylim for better visualization
plt.tight_layout()

# Show the plot
plt.show()
```

```python


# Assuming the 'dist' DataFrame is already defined
dist = dist[dist['year'] == '4Y-5Y']

# Create separate dataframes for attrited and existing customers
attrited_df = dist[dist['Customer_Status'] == 'Attrited Customer']
existing_df = dist[dist['Customer_Status'] == 'Existing Customer']

# Count the number of customers in each education category
attrited_counts = attrited_df['Gender'].value_counts()
existing_counts = existing_df['Gender'].value_counts()

# Combine the counts into a single DataFrame
df_education = pd.DataFrame({
    'Attrited_Customers': attrited_counts,
    'Existing_Customers': existing_counts
}).fillna(0)

# Calculate total customers and percentage of attrited customers
df_education['Total_Customers'] = df_education['Attrited_Customers'] + df_education['Existing_Customers']
df_education['Percentage_Attrited'] = (df_education['Attrited_Customers'] / df_education['Total_Customers']) * 100

# Define custom colors for each education category
colors = [
     'red',  'blue', 'green', 'purple'
]

# Plotting
plt.figure(figsize=(12, 8))
bars = plt.bar(df_education.index, df_education['Percentage_Attrited'], color=colors, edgecolor='black')

# Add percentage numbers above each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{yval:.2f}%', ha='center', va='bottom')

plt.xlabel('Gender')
plt.ylabel('Percentage of Attrited Customers (%)')
plt.title('Percentage of Attrited Customers by Education Level')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, max(df_education['Percentage_Attrited']) + 5)
plt.tight_layout()

# Show the plot
plt.show()
```
